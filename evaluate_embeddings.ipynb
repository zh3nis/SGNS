# Evaluate Saved SGNS Embeddings

This notebook reproduces the evaluation block from the end of `main.py` for a saved word2vec text file.

```python
import os
from gensim.models import KeyedVectors
from gensim.test.utils import datapath
```

```python
embedding_path = 'embeddings/sgns'
assert os.path.exists(embedding_path), f'Missing embedding file: {embedding_path}'
embedding_path
```

```python
wv_from_text = KeyedVectors.load_word2vec_format(embedding_path, binary=False)
ws353 = wv_from_text.evaluate_word_pairs(datapath('wordsim353.tsv'))
google = wv_from_text.evaluate_word_analogies(datapath('questions-words.txt'))
print(f'WS353 = {ws353[0][0]:.3f}, Google = {google[0]:.3f}')
```
